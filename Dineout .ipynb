Write a python program to scrape mentioned details from dineout.co.in

```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants')
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
restaurant = []
 
for i in soup.find_all ('div' ,class_="restnt-info cursor"):
    restaurant.append(i.text)
    
restaurant    
```

```python
location = []

for i in soup.find_all ('div' ,class_="restnt-loc ellipsis"):
    location.append(i.text)
    
location    
```

```python
cuisine = []

for i in soup.find_all('span' ,class_="double-line-ellipsis"):
    cuisine.append(i.text)
    
cuisine    
 
```

```python
rating = []

for i in soup.find_all ('div' ,class_="restnt-rating"):
    rating.append(i.text)
    
rating
```
